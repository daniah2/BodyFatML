```python
# 1Ô∏è. Import Required Libraries
# -------------------------------------------------------------
import pandas as pd
```

```python
# 2Ô∏è. Define File Paths
# (Change these paths to match your local file locations)
# -------------------------------------------------------------
file1_path = "BodyFat - Extended.csv"
file2_path = "bodyfat.csv.xls"     
file3_path = "bodyfat_dataset_v2.xlsx"
```

```python
# 3Ô∏è. Load the Datasets
# -------------------------------------------------------------
# Load first dataset (CSV)
file1 = pd.read_csv(file1_path)

# Try reading the second file as CSV first
try:
    file2 = pd.read_csv(file2_path)
except Exception:
    file2 = pd.read_excel(file2_path)

# Load third dataset (Excel)
file3 = pd.read_excel(file3_path)
```

```python
# 4Ô∏è. Define Common Columns
# -------------------------------------------------------------
common_columns = ["BodyFat", "Weight", "Height", "Chest", "Abdomen", "Hip", "Thigh", "Biceps"]
```

```python
# 5Ô∏è. Filter Datasets to Keep Only Common Columns
# -------------------------------------------------------------
df1 = file1[[col for col in common_columns if col in file1.columns]]
df2 = file2[[col for col in common_columns if col in file2.columns]]
df3 = file3[[col for col in common_columns if col in file3.columns]]
```

```python
# 6Ô∏è. Combine All Datasets
# -------------------------------------------------------------
combined_df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
# 7Ô∏è. Explore the Combined Dataset
# -------------------------------------------------------------
print("‚úÖ Combined Dataset Info:\n")
print(combined_df.info())
print("\nüìä First 5 Rows:\n")
print(combined_df.head())
```

```python
# 8Ô∏è. Save Combined Dataset to CSV
# -------------------------------------------------------------
output_file = "combined_bodyfat_dataset1.csv"
combined_df.to_csv(output_file, index=False)

print(f"\nüíæ Combined dataset saved as: {output_file}")
```
